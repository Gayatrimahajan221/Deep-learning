### image proccessing

```python
import cv2
```

### by default cv2 is going to read images as BGR

```python
imag1 = cv2.imread('CuteCat.jpeg')
imag1
```

```python
import matplotlib.pyplot as plt
plt.imshow(imag1)
```

### converting the image from bgr tot rgb

```python
img_org = cv2.cvtColor(imag1, cv2.COLOR_BGR2RGB)
img_org
```

```python
plt.imshow(img_org)
```

### seperate red , green , and blue channel

```python
r ,g, b = cv2.split(img_org)
```

```python
r
plt.imshow(r)
```

```python
plt.imshow(r , cmap='Reds')
```

```python
plt.imshow(g)
```

```python
plt.imshow(g , cmap = "Greens")
```

```python
b
```

```python
plt.imshow(b)
```

```python
plt.imshow(b, cmap='Blues')
```
